```python
plt.figure(figsize=(15,10))
sns.heatmap(X.corr(), annot=True,linewidths=0.5, cmap="Blues")
# Create a heatmap of the correlation matrix of the features of X
```

```python
# Split datasets into training and testing sets
from sklearn.model_selection import train_test_split
# 20% dataset will be used to test and 80% will be used to train 
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.20,random_state=1)

```

```python
from sklearn.svm import SVC  
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
prediction4=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
svm_acc = accuracy_score(Y_test, prediction4) #calculate the accuracy of classification model

print("Accuracy for with test dataset for Support Vector Machine:",svm_acc)
```

```python
#Evaluate the performance of the classification model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


cm_SVM=confusion_matrix(Y_test,prediction4) #calculates the confusion matrix for the prediction
sns.heatmap(cm_SVM, annot=True, fmt='d', cmap='Blues') # visualizing the confusion matrix using a heatmap
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
report = classification_report(Y_test, prediction4) #obtain the classification report
print("Classification Report:\n \n", report)
```
